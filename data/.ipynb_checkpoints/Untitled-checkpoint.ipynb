```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects = ('PrevPartyVoteShare', 'Polling', 'Raised', 'Spent', 'NameCommonness' , 'Incumbency', 'Wave', 'Gender', 'PresApproval', 'GovParty', 'LabourForce')
y_pos = np.arange(len(objects))
performance = [0.151,0.348,0.097, 0.087,0.0004,0.246,0.032,0.006,0.014,0.017,0.001]


plt.bar(y_pos, performance, align='center', alpha=0.7)
plt.xticks(y_pos, objects)

plt.xticks(rotation=70)

plt.ylabel('Random Forest Weight')
plt.ylabel('Features')
plt.title('Weights of Shortlisted Features')
 

plt.show()
```
